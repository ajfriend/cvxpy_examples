\begin{array}{ll}
\mbox{minimize} & \sum_{t=1}^\infty r^T q_t\\
\mbox{subject to} & q_{t+1} =  Af_{t+1} + q_t, \quad t = 0,1,2,\ldots\\
& 0 \leq q_t \leq Q, \quad t = 1,2,\ldots,
\end{array}


```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap
import cvxpy as cvx
from collections import defaultdict
from IPython.html import widgets
from itertools import izip

def mydraw(G,**kwds):
    kwdsOld = kwds.copy()
    
    if isinstance(kwds['node_shape'],str):
        kwds['node_shape'] = np.array([kwds['node_shape']]*len(kwds['nodelist']))
    for shape in set(kwds['node_shape']):
        inds = kwds['node_shape'] == shape
        nodes = [n for n, ind in izip(kwds['nodelist'], inds) if ind]
        
        kwds['nodelist'] = nodes
        kwds['node_size'] = kwds['node_size'][inds]
        kwds['node_shape'] = shape
        if not isinstance(kwds['node_color'],str):
            kwds['node_color'] = kwds['node_color'][inds]
        
        nx.draw(G,**kwds)
        kwds.update(kwdsOld)
    

def draw_graph(G,Q,r,q,f=None,ax=None):
    cmap = LinearSegmentedColormap.from_list('name', ['green', 'yellow', 'red'])
    if ax is None:
        fig = plt.figure(figsize=(6,6))
        ax = fig.add_subplot(111,aspect='equal')
    
    nodes = sorted(G.nodes())
    
    shapes = np.array(['o' if risk > 0 else 's' for risk in r])
    
    pos=nx.get_node_attributes(G,'pos') #dict keyed by node
    
    NODE_SIZE = 1000 #node plotting size
    EDGE_SIZE = 75
    
    q = np.maximum(q,0)
    s = max(Q[r > 0])
    Q = Q/s
    q = q/s
    
    s = Q[r == 0]
    q[r == 0] /= s
    Q[r == 0] = 1
    
    d = dict(nodelist=nodes,
             pos=pos,
             node_color='w',
             node_size=Q*NODE_SIZE,
             node_shape=shapes,
             cmap=cmap,
             linewidths=1,
             alpha=1,
             ax=ax,
             vmin=0, vmax=1)
    if f is not None:
        d['width'] = 1+np.abs(f)*EDGE_SIZE
    
    mydraw(G,**d)
    
    d['node_color'] = r
    d['alpha'] = .3
    mydraw(G,**d)
    
    d['node_size'] = q*NODE_SIZE
    d['alpha'] = 1
    d['linewidths'] = 2
    mydraw(G,**d)
    #nx.draw_networkx_nodes(G,**d)
    
    xmax = max([p[0] for p in pos.values()])
    xmin = min([p[0] for p in pos.values()])
    
    ymax = max([p[1] for p in pos.values()])
    ymin = min([p[1] for p in pos.values()])
    
    ax.axis([xmin-1,xmax+1,ymin-1,ymax+1])
    ax.invert_yaxis()
    
    
def opt_evac(G, Q, r, q0, T=20):
    nodes = sorted(G.nodes())
    A = nx.incidence_matrix(G,nodelist=nodes,oriented=True).toarray()
    n,m = A.shape

    Apos = np.fmax(A,0)
    Aneg = np.fmax(-A,0)

    q = defaultdict(lambda: cvx.Variable(n))
    f = defaultdict(lambda: cvx.Variable(m))
    q[0] = q0

    obj = 0
    constr = []
    for t in range(1,T+1):
        obj += q[t].T*r
        constr += [0 <= q[t], q[t] <= Q, q[t] == A*f[t] + q[t-1],
                   Aneg*cvx.pos(f[t]) + Apos*cvx.pos(-f[t]) <= q[t-1]]

    p = cvx.Problem(cvx.Minimize(obj), constr)
    p.solve(verbose=True)
    print "total risk: ", p.value

    f[0] = 0
    for var in [f, q]:
        for t in var:
            if t > 0:
                var[t] = np.array(var[t].value).flatten()
                
    return q, f

def draw_widget(G, Q, r, q, f):
    #todo: error at f[T+1]
    #todo: if any node size is < 0, get error.
    t=widgets.IntSliderWidget(min=0,max=len(q)-1,step=1,value=0)
    widgets.interact(lambda t: draw_graph(G,Q,r,q[t],f[t+1]),t=t);
    
def rand_data(G):
    for node in G.node:
        G.node[node]['pos'] = node
        
    n = len(G.nodes())
        
    q0 = np.random.rand(n)+.1
    q0[-1] = 0
    q0 = q0/sum(q0)

    Q = q0+np.random.rand(n)*np.mean(q0)*2
    r = np.fmin(np.random.rand(n) +.2,1)

    r[-1] = 0 # also set the cap to be big enough! oooh, what if the only safe spot is through a hot zone?
    Q[-1] = 1
    
    return G, Q, r, q0
```

```python
np.random.seed(0)
G = nx.grid_2d_graph(7,7)
G, Q, r, q0 = rand_data(G)
draw_graph(G,Q,r,q0)
```

```python
q, f = opt_evac(G, Q, r, q0, T=40)
draw_graph(G,Q,r,q[0],f[0])
```

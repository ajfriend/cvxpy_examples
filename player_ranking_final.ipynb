```python
import os  
import json
import datetime
from collections import defaultdict
from collections import namedtuple

class PlayerAppearance(object):
    __slots__= "name", "id", "minutes", "pm", 'pmpm',
    def __init__(self):
        for name in self.__slots__:
            setattr(self,name,None)
    def __str__(self):
        return str(self.items())
    def __repr__(self):
        return str(tuple(getattr(self,name) for name in self.__slots__))
    def items(self):
        return [(name, getattr(self,name)) for name in self.__slots__]

class Game(object):
    __slots__ = 'w','l','wscore','lscore','wplayers', 'lplayers','season',
    def __init__(self):
        for name in self.__slots__:
            setattr(self,name,None)
    def __str__(self):
        return str(self.items())
    def __repr__(self):
        return str(tuple(getattr(self,name) for name in self.__slots__))
    def items(self):
        return [(name, getattr(self,name)) for name in self.__slots__]

games = []

for fn in os.listdir('gamestats'):
     if fn.endswith('.json'):
        with open('gamestats/'+fn) as f:
            a = json.load(f)
            
            game = Game()
            game.w = str(a['resultSets'][1]['rowSet'][0][4])
            game.wscore = a['resultSets'][1]['rowSet'][0][21]
            
            game.l = str(a['resultSets'][1]['rowSet'][1][4])
            game.lscore = a['resultSets'][1]['rowSet'][1][21]
            
            game.season = int(a['resultSets'][0]['rowSet'][0][8])
            
            
            game.wplayers = []
            game.lplayers = []
            
            players = a['resultSets'][4]['rowSet']

            for player in players:
                if player[8] == None or player[8] == 0.0:
                    continue
                    
                playerapp = PlayerAppearance()
                
                        
                playerapp.id = player[4]
                playerapp.name = str(player[5])
                playerapp.pm = player[27]
                minutes = player[8].split(':')
                minutes = float(minutes[0]) + float(minutes[1])/60.0
                playerapp.minutes = minutes

                if player[2] == game.w:
                    game.wplayers.append(playerapp)
                elif player[2] == game.l:
                    game.lplayers.append(playerapp)
                else:
                    raise Exception("Player not on either team!")
                        

                # switch team order based on score
            if game.lscore > game.wscore:
                game.w, game.l, game.wscore, game.lscore, game.wplayers, game.lplayers = game.l, game.w, game.lscore, game.wscore, game.lplayers, game.wplayers
           
            games.append(game)
            
# sort out by season
#player_name2id = {v:k for k,v in player_id2name.iteritems()}
```

```python
# select just games in 2012 season
games = [game for game in games if game.season == 2012]

# filter out player appearances in games of only a few minutes
minminutes = 5.0
for game in games:
    game.wplayers = [player for player in game.wplayers if player.minutes >= minminutes]
    game.lplayers = [player for player in game.lplayers if player.minutes >= minminutes]
    
# compute total player playtime over all appearances
playerid2time = defaultdict(float)
for game in games:
    for player in game.wplayers + game.lplayers:
        playerid2time[player.id] += player.minutes

# filter out players who didn't play many minutes over the collection of games
minminutes = 1000.0
playerstoremove = {k for k,v in playerid2time.iteritems() if v < minminutes}
for game in games:
    game.wplayers = [player for player in game.wplayers if player.id not in playerstoremove]
    game.lplayers = [player for player in game.lplayers if player.id not in playerstoremove]
```

```python
%%time
import cvxpy as cvx
from collections import defaultdict
rpm = defaultdict(cvx.Variable) # 'real' plus minus

# fix 5x errors in team, player, and observed pm
# can find the right adjustment for player plus team 
# maybe normalize result by total game score, or divide sum by number of games
# do minutes for winner and loser, so we can remove people who don't play very many minutes
# in a given game, but do it so things still add up. also maybe remove players if they don't
# play many minutes overall

# i can actually compute the correct average for the *rest* of the team, excluding one player,
# based on the team average, total minutes, player average and player minutes
objective = 0
for game in games:
    wminutes = sum(p.minutes for p in game.wplayers)
    lminutes = sum(p.minutes for p in game.lplayers)
    wteampm = sum(rpm[p.id]*p.minutes for p in game.wplayers)/wminutes
    lteampm = sum(rpm[p.id]*p.minutes for p in game.lplayers)/lminutes
    
    for players, teamMinutes, teamPM, otherPM in [(game.wplayers, wminutes, wteampm, lteampm), (game.lplayers, lminutes, lteampm, wteampm) ]:
        objective += sum( cvx.abs( p.minutes*((teamMinutes - 5*p.minutes)*rpm[p.id]/(teamMinutes - p.minutes) + 4*teamMinutes*teamPM/(teamMinutes - p.minutes) - 5*otherPM) - p.pm ) for p in players )
    
    
    
    #todo: normalize objective by total number of appearances
    
    # model is just wrong. fix
    #(teamMinutes - 5*p.minutes)*rpm[p.id]/(teamMinutes - p.minutes) + 4*teamMinutes*teamPM/(teamMinutes - p.minutes)
    #(teamMinutes - 5*p.minutes)*rpm[p.id]/(teamMinutes - p.minutes) + 4*teamMinutes*teamPM/(teamMinutes - p.minutes)
    #objective += sum( cvx.square( p.minutes*((wminutes - 5*p.minutes)*rpm[p.id]/(wminutes - p.minutes) + 4*wminutes*wteampm/(wminutes - p.minutes)) - p.pm ) for p in game.wplayers )
    #objective += sum( cvx.square( p.minutes*(rpm[p.id] + 4*wteampm - 5*lteampm) - p.pm ) for p in game.wplayers )
    #objective += sum( cvx.square( p.minutes*(rpm[p.id] + 4*lteampm - 5*wteampm) - p.pm ) for p in game.lplayers )
    #objective += sum( cvx.square( p.minutes*((lminutes - 5*p.minutes)*rpm[p.id]/(lminutes - p.minutes) + 4*lminutes*lteampm/(lminutes - p.minutes)) - p.pm ) for p in game.lplayers )
    

#objective += sum( cvx.square(val) for val in rpm.values())
objective = cvx.Minimize(objective)

constraints = [sum(rpm.values()) == 0]

prob = cvx.Problem(objective, constraints)

result = prob.solve(solver=cvx.SCS, verbose=True)
```

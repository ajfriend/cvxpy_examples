```python
import os  
import json
import datetime

games = []
dates = []
#gameids = set()
for fn in os.listdir('gamestats'):
     if fn.endswith('.json'):
        with open('gamestats/'+fn) as f:
            a = json.load(f)
            name1 = str(a['resultSets'][1]['rowSet'][0][4])
            score1 = a['resultSets'][1]['rowSet'][0][21]
            name2 = str(a['resultSets'][1]['rowSet'][1][4])
            score2 = a['resultSets'][1]['rowSet'][1][21]
            #gameid = a['resultSets'][0]['rowSet'][0][2]
            gamedate = datetime.datetime.strptime(a['resultSets'][0]['rowSet'][0][0],'%Y-%m-%dT%H:%M:%S')
            season = a['resultSets'][0]['rowSet'][0][8]
            if season == '2012':
                dates.append(gamedate)
                #gameids.add(gameid)
                if score1 > score2:
                    games.append((name1,name2))
                else:
                    games.append((name2,name1))
                    
len(games)
```

```python
games
```

```python
import cvxpy as cvx
from collections import defaultdict
d = defaultdict(cvx.Variable)

objective = 0
for win, los in games:
    #objective += cvx.pos(d[los]-d[win]+1)
    objective += cvx.square(cvx.pos(d[los]-d[win]+1))
objective = cvx.Minimize(objective)

prob = cvx.Problem(objective)

result = prob.solve(solver=cvx.SCS, verbose=True)
```

```python
c = {team: d[team].value for team in d}
```

```python
import operator
teams, ranks = zip(*sorted(c.iteritems(), key=operator.itemgetter(1)))
pos = range(len(teams))
```

```python
plot(pos,ranks,'o')
xticks(pos, teams, rotation='vertical')
margins(0.1)
subplots_adjust(bottom=0.15)
```

```python
import os  
import json
import datetime
from collections import defaultdict
from collections import namedtuple

Game = namedtuple('Game', ['w','l','wscore','lscore','wplayers', 'lplayers','season'])
Player = namedtuple('Player', ['name','id','minutes', 'pm'])

games = []
player_id2name = {}
#player_id2time = defaultdict(int)

for fn in os.listdir('gamestats'):
     if fn.endswith('.json'):
        with open('gamestats/'+fn) as f:
            a = json.load(f)
            
            team1_abrv = str(a['resultSets'][1]['rowSet'][0][4])
            score1 = a['resultSets'][1]['rowSet'][0][21]
            team2_abrv = str(a['resultSets'][1]['rowSet'][1][4])
            score2 = a['resultSets'][1]['rowSet'][1][21]
            
            season = str(a['resultSets'][0]['rowSet'][0][8])
            
            
            team1_players = []
            team2_players = []
            
            if season == '2012':
                players = a['resultSets'][4]['rowSet']

                for player in players:
                    minutes = 0
                    if player[8] == None:
                        continue
                        
                    playerid = player[4]
                    playername = str(player[5])
                    plusminus = player[27]
                    playerteam = player[2]

                    player_id2name[playerid] = playername
                    minutes = player[8].split(':')
                    minutes = float(minutes[0]) + float(minutes[1])/60.0
                    

                    #player_id2time[player[4]] += minutes
                    player = Player(playername, playerid, minutes, plusminus)

                    if playerteam == team1_abrv:
                        team1_players.append(player)
                    elif playerteam == team2_abrv:
                        team2_players.append(player)
                    else:
                        raise Exception("Player not on either team!")
                        

                if score1 > score2:
                    games.append(Game(team1_abrv, team2_abrv, score1, score2, team1_players, team2_players, season))
                else:
                    games.append(Game(team2_abrv, team1_abrv, score2, score1, team2_players, team1_players, season))
                    
player_name2id = {v:k for k,v in player_id2name.iteritems()}
len(games)
```

```python
import cvxpy as cvx
from collections import defaultdict
d = defaultdict(cvx.Variable)

objective = 0
for winner, loser in games:
    namew, scorew, playersw = winner
    namel, scorel, playersl = loser
    
    winner = sum(d[player[0]]*player[1] for player in playersw)
    loser = sum(d[player[0]]*player[1] for player in playersl)
    objective += cvx.pos(loser-winner+1)
    
objective += sum(player for player in d.values())
objective = cvx.Minimize(objective)

constraints = [player >= 0 for player in d.values()]

prob = cvx.Problem(objective, constraints)

result = prob.solve(solver=cvx.ECOS, verbose=True)
```

```python
def evalrank(teamrank, games):
    count = 0
    for winner, loser in games:
        if teamrank[winner[0]] > teamrank[loser[0]]:
            count += 1
    
    return float(count)/len(games)
```

```python
from scipy import misc
import matplotlib.pyplot as plt
import numpy as np

l = misc.lena()
l = l.astype(numpy.float64, copy=False)
l = l/np.max(l) #rescale pixels into [0,1]

plt.imshow(l, cmap=plt.cm.gray)
plt.show()
```

```python
from PIL import Image, ImageDraw

num_lines = 10
width = 10
imshape = l.shape

def drawRandLine(draw,width):
    x = [np.random.randint(0,im.size[0]) for i in range(2)]
    y = [np.random.randint(0,im.size[1]) for i in range(2)]
    xy = zip(x,y)
    #fill gives the color
    draw.line(xy,fill=255,width=width)

im = Image.new("L",imshape)
draw = ImageDraw.Draw(im)
for i in range(num_lines):
    drawRandLine(draw,width)
del draw
im.show()
```

```python
err = np.asarray(im,dtype=np.bool)
r = l.copy()
r[err] = 1.0
plt.imshow(r, cmap=plt.cm.gray)
```

```python
import itertools
idx2pair = np.nonzero(err)
idx2pair = zip(idx2pair[0].tolist(), idx2pair[1].tolist())
pair2idx = dict(itertools.izip(idx2pair, xrange(len(idx2pair))))
idx2pair = np.array(idx2pair) #convert back to numpy array
```

```python
r[idx2pair[:,0],idx2pair[:,1]] #should be all ones
```

```python

# worry about signes of the b term

def addVarPairy(pair): ## just for Gy
    idx = pair2idx[pair]
    
    # first, handle this pixel's gradient
    if pair[0] < r.shape[0]-1: #interior pixel
        down = (pair[0]+1,pair[1])
        if down in pair2idx: #lower is variable
            #Dy.append((idx, pair2idx[(pair[0]+1,pair[1])]))
            #by.append(0)
            return (idx, pair2idx[down]), 0
        else: #lower is constant
            #Dy.append((idx,))
            #by.append(r[(pair[0]+1,pair[1])])
            return (idx,), r[down]
    else: #boundary pixel
        #Dy.append(())
        #by.append(0)
        return (), 0
    
def addVarPairx(pair): ## just for Gx
    idx = pair2idx[pair]
    
    if pair[1] < r.shape[1]-1: #interior pixel
        right = (pair[0],pair[1]+1)
        if right in pair2idx: #right is variable
            return (idx, pair2idx[right]), 0
        else: #lower is constant
            return (idx,), r[right]
    else: #boundary pixel
        return (), 0
    
def addFixPairy(pair): ## just for Gy
    #this one is fixed,but below could be variable, fixed, or past boundary
    
    if pair[0] < r.shape[0]-1: #interior pixel
        down = (pair[0]+1,pair[1])
        if down in pair2idx: #down is variable
            idx = pair2idx[down]
            return (idx,), r[pair]
        else: #down is fixed
            return (), r[pair] - r[down]
    else: #boundary pixel
        return (), 0
    

def addFixPairx(pair): ## just for Gx
    #this one is fixed, but right could be variable, fixed, or past boundary
    
    if pair[1] < r.shape[1]-1: #interior pixel
        right = (pair[0],pair[1]+1)
        if right in pair2idx: #right is variable
            idx = pair2idx[right]
            return (idx,), r[pair]
        else: #right is fixed
            return (), r[pair] - r[right]
    else: #boundary pixel
        return (), 0
    

def addPair(pair):
    #return gxinfo, gyinfo
    if pair in pair2idx:
        gyinfo = addVarPairy(pair)
        gxinfo = addVarPairx(pair)
    else:
        gyinfo = addFixPairy(pair)
        gxinfo = addFixPairx(pair)
    return gyinfo, gxinfo




# input should be a set of pair tuples, allows us to loop and check for membership
def getginfo(idx2pair):
    for pair in idx2pair: #loop through unknown pixels
        pair = tuple(pair)
        yield addPair(pair) # add this pixel
        
        left = (pair[0],pair[1]-1)
        if pair[1] > 0 and left not in pair2idx: #if not on left boundary, add left pixel
            yield addPair(left)
        
        top = (pair[0]-1,pair[1])
        topright = (pair[0]-1,pair[1]+1)
        #if not on top boundary, top is fixed, and top not already toched by upper right pixel
        if pair[0] > 0 and top not in pair2idx and topright not in pair2idx:
            yield addPair(top)
            

Dx = [] #idx1, idx2 tuples
Dy = []
bx = []
by = []

for gyinfo, gxinfo in getginfo(idx2pair):
    Dy.append(gyinfo[0])
    by.append(gyinfo[1])
    
    Dx.append(gxinfo[0])
    bx.append(gxinfo[1])   
```

```python
import scipy.sparse as sp
from cvxopt import spmatrix

m = len(Dx)
n = idx2pair.shape[0]

Ix = []
Jx = []
Vx = []

Iy = []
Jy = []
Vy = []

for row in xrange(m):
    if len(Dx[row]) == 2:
        i,j = Dx[row]
        Ix.append(row)
        Jx.append(i)
        Vx.append(1)
        
        Ix.append(row)
        Jx.append(j)
        Vx.append(-1)
    elif len(Dx[row]) == 1:
        i, = Dx[row]
        Ix.append(row)
        Jx.append(i)
        Vx.append(1)
    else:
        pass
    
    if len(Dy[row]) == 2:
        i,j = Dy[row]
        Iy.append(row)
        Jy.append(i)
        Vy.append(1)
        
        Iy.append(row)
        Jy.append(j)
        Vy.append(-1)
    elif len(Dy[row]) == 1:
        i, = Dy[row]
        Iy.append(row)
        Jy.append(i)
        Vy.append(1)
    else:
        pass
    
#Gx = sp.coo_matrix((Vx,(Ix,Jx)),shape=(m,n))
Gx = spmatrix(Vx, Ix, Jx,(m,n))

#Gy = sp.coo_matrix((Vy,(Iy,Jy)),shape=(m,n))
Gy = spmatrix(Vy, Iy, Jy,(m,n))

bx = np.array(bx)
by = np.array(by)
```

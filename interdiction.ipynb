```python
#%config InlineBackend.figure_format = 'pdf'
#%config InlineBackend.figure_format = 'svg'
%config InlineBackend.figure_format = 'retina'

import matplotlib.pyplot as plt
import numpy as np
import networkx as nx
import cvxpy as cvx

from matplotlib import rcParams
rcParams.update({'font.size': 16})

def showPaths(G,pos,p,x=None):
    ''' Takes directed graph G, node positions pos, and edge probabilities p
        optionally takes x to plot a path
    '''
    n = G.number_of_nodes()
    nx.draw_networkx_nodes(G,pos,nodelist=range(1,n-1),node_color='w',node_size=100, with_labels=False)
    
    nx.draw_networkx_nodes(G,pos,
                       nodelist=[0],
                       node_color='w',
                       node_size=300)
    nx.draw_networkx_nodes(G,pos,
                       nodelist=[0],
                       node_color='g',
                       node_size=300,alpha=.2)
    nx.draw_networkx_nodes(G,pos,
                       nodelist=[n-1],
                       node_color='w',
                       node_size=300)
    nx.draw_networkx_nodes(G,pos,
                       nodelist=[n-1],
                       node_color='r',
                       node_size=300,alpha=.2)
    
    if x is None:
        alpha = 1
    else:
        alpha = .15
        
    edges = nx.draw_networkx_edges(G,pos,edge_color=p,width=4,
                                   edge_cmap=plt.cm.RdYlGn,arrows=False,alpha=alpha)
    
    if x is not None:
        path_inds = list(np.flatnonzero(x > .1))
        path_p = [p[i] for i in path_inds]
        path_pairs = [G.edges()[i] for i in path_inds]
        edges = nx.draw_networkx_edges(G,pos,edge_color=path_p,width=8,
                                       edge_cmap=plt.cm.RdYlGn,edgelist=path_pairs,arrows=False)
        
    plt.colorbar(edges,label='Evasion probability')

    fig = plt.gcf()
    ax = plt.gca()
    fig.set_size_inches(8,6)
    ax.axis([-0.02,.97,-0.02,.98])

    plt.axis('off')
    
def optPath(G,p,edge2ind):
    ''' Find the optimal smuggler's path in directed graph G
        with edge evasion probabilities p and dictionary
        edge2ind bringing node pairs to edge indices
    '''
    H = G.copy()
    p = np.minimum(p,1)
    w = -np.log(p)
    
    for i in H:
        for j in H[i]:
            H[i][j]['w'] = w[edge2ind[(i,j)]]
            
    path = nx.shortest_path(H,0,n-1,'w')
            
    foo = [edge2ind[(i,j)] for i,j in zip(path[:-1],path[1:])]
    x = np.zeros_like(p)
    x[foo] = 1
    return x
```

# Problem statement

A smuggler moves along a directed graph (cycles are allowed) with $n$ nodes and $m$ edges from node $0$ (the source) to node $n-1$ (the destination). Edge $j$ has evastion probability $p_j$, which is the probability that the smuggler moves along that edge undetected. The detection events on the edges are independent, so the probability that the smuggler makes it to the destination undetected is
$\prod_{j \in \mathcal{P}} p_j$, where $\mathcal{P} \subset \lbrace 0,\ldots, m-1\rbrace$ is the set of edges on the smuggler's path. We assume that the smuggler knows the evasion probabilities and will take a path that maximizes the probability of making it to the destination node undetected. We let $P^\mathrm{max}$ denote this maximum probability (over paths). Note that this is a function of the edge detection failure probabilities.

The evasion probability on an edge depends on the interdiction resources allocated to the edge. Here we will use a very simple model, with $r_j \in \mathbf{R}_+$ denoting the effort (say, yearly budget) allocated to edge $j$, with associated evasion probability $p_j = e^{-a_j r_j}$ , where $a_j \in \mathbf{R}_{++}$ are given. The resources will be constrained in some way such that $r \in \mathcal{C}$, where $\mathcal{C}$ represents some resource allocation constraints such as, say, a maximum for each edge or a total budget constraint.

The problem is to choose resources $r \in \mathbf{R}^m$, subject to the constraints, so as to minimize $P^\mathrm{max}$.

\\[
\begin{array}{ll}
  \mbox{minimize} & P^\mathrm{max}(r) \\
  \mbox{subject to} & r \in \mathcal{C}
\end{array}
\\]


# Smuggler's perspective

To solve the problem, we first consider the smuggler's objective and formulate it as an optimization problem. Since the smuggler chooses a path only after the resources are allocated, we will temporarily drop the resource variable $r$ and consider the evasion probabilities $p$ as fixed. Note that the smuggler's problem can be cast and solved exactly as a [shortest path problem in graph theory](http://en.wikipedia.org/wiki/Shortest_path_problem). We formulate it as a convex problem so that we can use the model to thwart the smuggler.

Let $x \in \lbrace 0,1 \rbrace^m$ be a decision variable where $x_j = 1$ denotes that the smuggler's path includes edge $j$. The $x$ variable needs to satisfy certain constraints to represent a valid path. Firstly, the number of times the path exits the source node must be exactly one more than the number of times it exits the source node. Similarly, the number of times the path enters the destination node must be exactly one more than the number of times it exits destination node. For all other nodes in the graph, the number of times the path enters and exits the node must be equal. These constraints can be represented as $Ax = b$, where

\\[
b_i = \begin{cases} 
-1 & i = 0 \\
+1 & i = n-1\\
0 & \mbox{otherwise,}
\end{cases}
\\]

\\[
A_{ij} = 
\begin{cases} 
+1 & \mbox{if edge $j$ enters node $i$} \\
-1 & \mbox{if edge $j$ exits node $i$} \\
0 & \mbox{otherwise.}
\end{cases}
\\]

Since the smuggler's objective $\prod_{j \in \mathcal{P}} p_j$ is positive, it is equivalent for him to maximize the logarithm of the objective, which we can write as 

\\[
\begin{align}
\log\left(\prod_{j \in \mathcal{P}} p_j\right) &= \sum_{j \in \mathcal{P}} \log(p_j)\\
&= \sum_{j=0}^{n-1} c_j x_j,
\end{align}
\\]
where we've assigned $c_j = \log(p_j)$.


The smuggler's problem can be written as

\\[
\begin{array}{ll}
  \mbox{maximize} & c^T x \\
  \mbox{subject to} & A x = b\\
  & x \in \lbrace 0,1 \rbrace^m.
\end{array}
\\]

This is a linear problem with Boolean variables, which is not convex, but we can transform it into a convex problem by relaxing the Boolean constraints to get

\\[
\begin{array}{ll}
  \mbox{maximize} & c^T x \\
  \mbox{subject to} & A x = b\\
  & x \geq 0.
\end{array}
\\]

It turns out that this convex problem solves the original Boolean problem *exactly*. That is, the optimal value of the Boolean and relaxed LP problem are the same. In addition the solution to the LP, $x^\star$ will be Boolean when there is a unique optimal path. In the presence of multiple optimal paths, the $x^\star$ may have fractional entries, but an optimal path can still be recovered.

# Smuggler's path example
We'll solve the smuggler's problem on an example network to see an example of an optimal path.

We'll create an $N \times N$ grid of points with small perturbations to make the graph irregular, connecting two nodes with two directed edges (one going each way) if the nodes are within distance $\eta$ of each other. Pairs of edges $(i,j)$ and $(j,i)$ will have identical evasion probabilities, i.e., the evasion probability is the same in either direction between two connected nodes. The evasion probability will be distributed like $p_k = e^{-a_k}$, where $a_k$ is a uniform random variable over the interval $\left[0,1\right]$.

The smuggler will find an optimal path from node $0$ in the upper-left corner of the graph plot to node $n-1$ in the lower-right corner of the graph plot.

```python
np.random.seed(3)

N = 10 # n = N^2 nodes
mu = .045 # node placement noise
eta = 0.13 # edge radius
n = N*N

# generate perterbed grid positions for the nodes
pos = [(float(i)/N + mu*np.random.rand(), float(j)/N + mu*np.random.rand())\
       for i in range(N) for j in range(N-1,-1,-1)]
pos = dict(enumerate(pos))
# connect nodes with edges
G = nx.random_geometric_graph(n,eta,pos=pos)

# assign a random a_j to edge j in graph G
for i in G:
    for j in G[i]:
        if i > j:
            G[i][j]['a'] = np.random.rand()
            
# create directed graph H from G, with both directions inheriting a_j from G
H = nx.DiGraph(G)
a = np.array([e[2]['a'] for e in H.edges_iter(data=True)])
p = np.exp(-a)
for u,v,d in H.edges(data=True):
    d.clear()
del G

A = nx.incidence_matrix(H,oriented=True).toarray()
n,m = A.shape         
      


edge2ind = {e: i for i,e in enumerate(H.edges_iter())}

B = np.zeros((m/2,m))
count = 0
for i in H:
    for j in H[i]:
        if i < j:
            B[count,edge2ind[(i,j)]] = 1
            B[count,edge2ind[(j,i)]] = -1
            count += 1
            
showPaths(H,pos,p)
```

We form the smuggler's relaxed convex problem and solve it to find his optimal path. We plot the path below.

```python
b = np.zeros(n)
b[0] = -1
b[n-1] = 1
c = -a
x = cvx.Variable(m)
constr = [A*x == b,x>=0, x <= 1]
cvx.Problem(cvx.Maximize(x.T*c),constr).solve(verbose=True)
x = np.array(x.value).flatten()
```

```python
showPaths(H,pos,p,x)
```

The evasion probability of the smuggler's optimal path is about 1.7%

```python
np.exp(x.dot(c))
```

We run a discrete graph-theoretic shortest path algorithm on the same graph to check that we get the same optimal path.

```python
y = optPath(H,p,edge2ind)
showPaths(H,pos,p,y)
np.exp(y.dot(c))
```

# Security's perspective

We return to the main problem, which is to allocate interdiction resources to the edges to make it as likely as possible to detect the smuggler along his optimal path.

We model the edge evasion probability as a function of the resources spent on that edge as $p_j = e^{-a_j r_j}$. This gives us $c_j = \log(p_j) = -a_j r_j$, where the resources spent on the edge $r_j \geq 0$. As the $a_j$ vary, some edges will be easier to secure than others.

Taking the dual of the smuggler's problem allows us to minimize the evasion probability of the smuggler, who is trying to maximize it. The following is the dual of the smuggler's primal problem. The primal and dual problems have equal optimal values.

\\[
\begin{array}{ll}
  \mbox{minimize} & b^T \lambda \\
  \mbox{subject to} & E^T \lambda \geq c
\end{array}
\\]

Since $c_j = -a_j r_j$, we can add in the resource variables, with approprate constraints on the resources, represented by the set $\mathcal{C}$.

\\[
\begin{array}{ll}
  \mbox{minimize} & b^T \lambda \\
  \mbox{subject to} & E^T \lambda \geq c \\
  & c_i = a_i r_i \\
  & r \in \mathcal{C}
\end{array}
\\]

Solving the previous problem will allocate resources to the edges to maximize the chance of detecting the smuggler, who will take the path which minimizes that chance.

# Security example

We'll use the same graph as above, with the same values for $a_j$. That is, $p_j = e^{-a_j r_j}$. The previous example took the specific allocation of $r_j = 1$ uniformly.

In this example, we'll see how much we can improve over the uniform allocation, using the same total resources. The resource limit will be given by $\sum_{j=0}^{m-1} r_j = m$. We'll also limit the total resources that can be spent on any single edge by $r_j \leq 5$. We'll also constrain that the evasion probability is equal in both directions. That is edge $(i,j)$ and edge $(j,i)$ have equal evasion probabilities. We'll enforce that constraint with $Br = 0$, for some matrix $B$.

The final model is
\\[
\begin{array}{ll}
  \mbox{minimize} & b^T \lambda \\
  \mbox{subject to} & E^T \lambda \geq c \\
  & c_j = a_j r_j,\ r_j \geq 0,\ r_j \leq 5\quad \forall i \\
  & Br = 0 \\
  &\sum_{j=0}^{m-1} r_j = m
\end{array}
\\]

We solve the model below and report the evasion probability of the smuggler's optimal path.

```python
nu = cvx.Variable(n)
r = cvx.Variable(m)
constr = [A.T*nu >= -cvx.mul_elemwise(a,r), cvx.sum_entries(r) == m, r >= 0, B*r == 0, r <= 5]
cvx.Problem(cvx.Minimize(nu.T*b),constr).solve(verbose=True)
nu = np.array(nu.value).flatten()
r = np.array(r.value).flatten()
```

The smuggler is about 9000 times more likely to be detected with the optimal allocation, as compared with the initial uniform allocation of resources.

```python
np.exp(x.dot(c))/np.exp(nu.dot(b))
```

Here are the resulting edge evasion probabilities from the optimal allocation.

```python
p = np.exp(-a*r)
showPaths(H,pos,p)
```

We can solve the smuggler's primal problem, but we won't recover a Boolean solution for $x^{\star}$ because the optimal path is not unique. Note, however, that the optimal evasion probability is the same.

```python
c = np.log(p)
x = cvx.Variable(m)
constr = [A*x == b,x>=0, x <= 1]
cvx.Problem(cvx.Maximize(x.T*c),constr).solve(verbose=True)
x = np.array(x.value).flatten()

plt.plot(x)
np.exp(x.dot(c))
```

We use a discrete shortest path algorithm from graph theory to recover an optimal path for the smuggler in the case that the path is not unique and plot it below. We again check that the detection probability is what we predicted previously.

```python
x = optPath(H,p,edge2ind)
showPaths(H,pos,p,x)
```

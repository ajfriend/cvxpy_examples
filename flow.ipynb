The bad guy tries to solve the problem

\\[
\begin{array}{ll}
  \mbox{maximize} & c^T x \\
  \mbox{subject to} & E x = b\\
  & x \geq 0,
\end{array}
\\]
where $c_i = \log(p_i)$ and $p_i$ is the probability that the bad guy gets through the edge $j$ without being detected. The optimal value is the probability that the bad guy gets through, taking the optimal path. $x_i = 1$ means he takes edge $i$, $x_i = 0$ means he does not take that edge. $E$ and $b$ represent conservation laws to ensure that only a single path is chosen.

Taking $p_i = e^{-a_j r_j}$ is a reasonable model, forcing $c_j = -a_j r_j$ with $r_j \geq 0$.

The graph is going to be a directed graph. The path will be from a source to a sink in the graph. We don't necessarily need to double the edges in the regular graph, but doing so keeps from having to choose an arbitrary orientation.

The following dual has the same optimal value as the primal problem. Solving the dual allows us to minimize the best possible path for the bad guy.

\\[
\begin{array}{ll}
  \mbox{minimize} & b^T \lambda \\
  \mbox{subject to} & E^T \lambda \geq c
\end{array}
\\]

The matrix $E$ is nodes by edges with a +1 for the input node and -1 for the output node.

```python
import networkx as nx
import matplotlib.pyplot as plt
np.random.seed(0)

n=10
noise = .045

pos = [(float(i)/n + noise*np.random.rand(1)[0] ,float(j)/n + noise*np.random.rand(1)[0]) for i in range(n) for j in range(n)]
pos = dict(enumerate(pos))
G = nx.random_geometric_graph(n*n,0.12,pos=pos)
m = G.number_of_edges()

for i in G:
    for j in G[i]:
        if i > j:
            a = np.random.rand(1)[0]
            G[i][j]['a'] = a
            G[i][j]['prob'] = np.exp(-a)
            
probs = [e[2]['prob'] for e in G.edges_iter(data=True)]

nodes = nx.draw_networkx_nodes(G,pos,node_color='w',node_size=100, with_labels=False)
edges = nx.draw_networkx_edges(G,pos,edge_color=probs,width=4,
                               edge_cmap=plt.cm.RdYlGn)
plt.colorbar(edges)
plt.axis('off')

H=nx.DiGraph(G)
```

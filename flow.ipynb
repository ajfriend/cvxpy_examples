# Problem statement

A bad guy moves along a directed graph (cycles are allowed) with $n$ nodes and $m$ edges from node $0$ (the source) to node $n-1$ (the destination). Edge $j$ has evastion probability $p_j$, which is the probability that the bad guy moves along that edge undetected. The detection events on the edges are independent, so the probability that the bad guy makes it to the destination undetected is
$\prod_{j \in \mathcal{P}} p_j$, where $\mathcal{P} \subset \lbrace 0,\ldots, m-1\rbrace$ is the set of edges on the bad guy's path. We assume that the bad guy knows the evasion probabilities and will take a path that maximizes the probability of making it to the destination node undetected. We let $P^\mathrm{max}$ denote this maximum probability (over paths). Note that this is a function of the edge detection failure probabilities.

The evasion probability on an edge depends on the interdiction resources allocated to the edge. Here we will use a very simple model, with $r_j \in \mathbf{R}_+$ denoting the effort (say, yearly budget) allocated to edge $j$, with associated evasion probability $p_j = e^{-a_j r_j}$ , where $a_j \in \mathbf{R}_{++}$ are given. The resources will be constrained in some way such that $r \in \mathcal{C}$, where $\mathcal{C}$ represents some resource allocation constraints such as, say, a maximum for each edge or a total budget constraint.

The problem is to choose resources $r \in \mathbf{R}^m$, subject to the constraints, so as to minimize $P^\mathrm{max}$.

\\[
\begin{array}{ll}
  \mbox{minimize} & P^\mathrm{max}(r) \\
  \mbox{subject to} & r \in \mathcal{C}
\end{array}
\\]


# Bad guy's perspective

To solve the problem, we first consider the bad guy's objective and formulate it as an optimization problem. Since the bad guy chooses a path only after the resources are allocated, we will temporarily drop the resource variable $r$ and consider the evasion probabilities $p$ as fixed.

Let $x \in \lbrace 0,1 \rbrace^m$ be a decision variable where $x_j = 1$ denotes that the bad guy's path includes edge $j$. The $x$ variable needs to satisfy certain constraints to represent a valid path. Firstly, the number of times the path exits the source node must be exactly one more than the number of times it exits the source node. Similarly, the number of times the path enters the destination node must be exactly one more than the number of times it exits destination node. For all other nodes in the graph, the number of times the path enters and exits the node must be equal. These constraints can be represented as $Ax = b$, where

\\[
b_i = \begin{cases} 
-1 & i = 0 \\
+1 & i = n-1\\
0 & \mbox{otherwise,}
\end{cases}
\\]

\\[
A_{ij} = 
\begin{cases} 
+1 & \mbox{if edge $j$ enters node $i$} \\
-1 & \mbox{if edge $j$ exits node $i$} \\
0 & \mbox{otherwise.}
\end{cases}
\\]

Since the bad guy's objective $\prod_{j \in \mathcal{P}} p_j$ is positive, it is equivalent for him to maximize the logarithm of the objective, which we can write as 

\\[
\begin{align}
\log\left(\prod_{j \in \mathcal{P}} p_j\right) &= \sum_{j \in \mathcal{P}} \log(p_j)\\
&= \sum_{j=0}^{n-1} c_j x_j,
\end{align}
\\]
where we've assigned $c_j = \log(p_j)$.


The bad guy's problem can be written as

\\[
\begin{array}{ll}
  \mbox{maximize} & c^T x \\
  \mbox{subject to} & A x = b\\
  & x \in \lbrace 0,1 \rbrace^m.
\end{array}
\\]

This is a linear problem with Boolean variables, which is not convex, but we can transform it into a convex problem by relaxing the Boolean constraints to get

\\[
\begin{array}{ll}
  \mbox{maximize} & c^T x \\
  \mbox{subject to} & A x = b\\
  & x \geq 0.
\end{array}
\\]

It turns out that this convex problem solves the original Boolean problem *exactly* with $x^\star \in \lbrace 0,1 \rbrace^m$. (XXX: This is the case only when there is a unique optimal path. In the presence of multiple optimal paths, the $x^\star$ may have fractional entries, but an optimal path can still be recovered.)

# Smuggler's path example
We'll solve the smuggler's problem on an example network to see an example of an optimal path.

We'll create an $N \times N$ grid of points with small perturbations to make the graph irregular, connecting two nodes with two directed edges (one going each way) if the nodes are within distance $\eta$ of each other. Pairs of edges $(i,j)$ and $(j,i)$ will have identical evasion probabilities, i.e., the evasion probability is the same in either direction between two connected nodes. The evasion probability will be distributed like $p_k = e^{-a_k}$, where $a_k$ is a uniform random variable over the interval $\left[0,1\right]$.

The smuggler will find an optimal path from node $0$ in the upper-left corner of the graph plot to node $n-1$ in the lower-right corner of the graph plot.

```python
np.random.seed(0)

N = 10 # n = N^2 nodes
mu = .045 # node placement noise
eta = 0.12 # edge radius
n = N*N

pos = [(float(i)/N + mu*np.random.rand(), float(j)/N + mu*np.random.rand())\
       for i in range(N) for j in range(N-1,-1,-1)]
pos = dict(enumerate(pos))
G = nx.random_geometric_graph(n,eta,pos=pos)

# assign a random a_j to edge j in graph G
for i in G:
    for j in G[i]:
        if i > j:
            G[i][j]['a'] = np.random.rand()
            
# create directed graph H from G, with both directions inheriting a_j from G
H = nx.DiGraph(G)
A = nx.incidence_matrix(H,oriented=True).toarray()
n,m = A.shape         
      
a = np.array([e[2]['a'] for e in H.edges_iter(data=True)])
p = np.exp(-a)


edges = nx.draw_networkx_edges(H,pos,edge_color=p,width=4,
                               edge_cmap=plt.cm.RdYlGn,arrows=False)
nodes = nx.draw_networkx_nodes(H,pos,node_color='w',node_size=100)


fig = plt.gcf()
ax = plt.gca()
fig.set_size_inches(8,6)
ax.axis([-0.02,.97,-0.02,.98])
plt.colorbar(edges)
plt.axis('off')
```

```python
b = np.zeros(n)
b[0] = -1
b[n-1] = 1
c = np.array([-e[2]['a'] for e in H.edges_iter(data=True)])
```

```python
import cvxpy as cvx
x = cvx.Variable(m)
constr = [E*x == b,x>=0, x <= 1]
cvx.Problem(cvx.Maximize(x.T*c),constr).solve(verbose=True)
x = np.array(x.value).flatten()
```

```python
path = list(np.flatnonzero(x > .5))
p_path = [p[e] for e in path]

edges_path = [H.edges()[e] for e in path]

#nx.draw(H,with_labels=True,pos=pos,edgelist=edges_path,node_size=100)

nodes = nx.draw_networkx_nodes(H,pos,node_color='w',node_size=100, with_labels=False)

edges = nx.draw_networkx_edges(H,pos,edge_color=p,width=4,
                               edge_cmap=plt.cm.RdYlGn,arrows=False,alpha=.15)

edges = nx.draw_networkx_edges(H,pos,edge_color=p_path,width=8,
                               edge_cmap=plt.cm.RdYlGn,edgelist=edges_path,arrows=False)
plt.colorbar(edges,label='Evasion probability')

# nx.draw_networkx_nodes(H,pos,
#                        nodelist=[0],
#                        node_color='g',
#                        node_size=300)

# nx.draw_networkx_nodes(H,pos,
#                        nodelist=[n-1],
#                        node_color='r',
#                        node_size=300)

# nx.draw_networkx_labels(H,pos,{0:'s',n-1:'t'},font_size=16)

fig = plt.gcf()
ax = plt.gca()
fig.set_size_inches(8,6)
ax.axis([-0.02,.97,-0.02,.98])

plt.axis('off')
```

Taking $p_i = e^{-a_j r_j}$ is a reasonable model, forcing $c_j = -a_j r_j$ with $r_j \geq 0$.

The graph is going to be a directed graph. The path will be from a source to a sink in the graph. We don't necessarily need to double the edges in the regular graph, but doing so keeps from having to choose an arbitrary orientation.

The following dual has the same optimal value as the primal problem. Solving the dual allows us to minimize the best possible path for the bad guy.

\\[
\begin{array}{ll}
  \mbox{minimize} & b^T \lambda \\
  \mbox{subject to} & E^T \lambda \geq c
\end{array}
\\]

The matrix $E$ is nodes by edges with a +1 for the input node and -1 for the output node.

```python
a = np.array([e[2]['a'] for e in H.edges_iter(data=True)])
import cvxpy as cvx
nu = cvx.Variable(n)
r = cvx.Variable(m)
constr = [E.T*nu >= -cvx.mul_elemwise(a,r), r==1]#cvx.sum_entries(r) == n, r >= 0]
cvx.Problem(cvx.Minimize(nu.T*b),constr).solve(verbose=True)
nu = np.array(nu.value).flatten()
r = np.array(r.value).flatten()
```
